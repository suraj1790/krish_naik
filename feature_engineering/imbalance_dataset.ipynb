## Handling imbalanced dataset part 1

1. up sampling
2. down sampling

```python
import numpy as np
import pandas as pd

np.random.seed(123)

n_sample = 1000
sample_ratio = 0.9
n_class_0 = int(n_sample * sample_ratio)
n_class_1 = int(n_sample - class_0)
```

```python
n_class_0,n_class_1
```

```python
# create our datafraome with imbalance dataset

class_0 = pd.DataFrame({
    "feature_1":np.random.normal(loc=0,scale=1,size=n_class_0),
    "feature_2":np.random.normal(loc=0,scale=1,size=n_class_0),
    "target" : [0]*n_class_0
})

class_1 =  pd.DataFrame({
    "feature_1":np.random.normal(loc=2,scale=1,size=n_class_1),
    "feature_2":np.random.normal(loc=2,scale=1,size=n_class_1),
    "target" : [1]*n_class_1
})
```

```python
class_0
```

```python
class_1
```

```python
df = pd.concat([class_0,class_1])
```

```python
df
```

```python
df.shape
```

```python
df.info()
```

```python
df["target"].value_counts()
```

```python
df_minaroty = df[df["target"]==1]
df_majority = df[df["target"]==0]
```

```python
from sklearn.utils import resample
```

```python
df_minority_upsample = resample(df_minaroty,replace=True,n_samples=len(df_majority),random_state=42)
df_minority_upsample
```

```python
df_unsample = pd.concat([df_majority,df_minority_upsample])
```

```python
df_unsample
```

```python
df_unsample["target"].value_counts()
```

### same as Down sampling -- reduce the majority value

```python
np.random.seed(123)

n_samples = 1000
class_ratio = 0.9
n_class_0 = int(n_samples * class_ratio)
n_class_1 = n_samples - n_class_0

class_0 = pd.DataFrame({
    "feature_1":np.random.normal(loc=0,scale=1,size=n_class_0),
    "feature_2":np.random.normal(loc=0,scale=1,size=n_class_0),
    "target" : [0]*n_class_0
})

class_1 =  pd.DataFrame({
    "feature_1":np.random.normal(loc=2,scale=1,size=n_class_1),
    "feature_2":np.random.normal(loc=2,scale=1,size=n_class_1),
    "target" : [1]*n_class_1
})

df2 = pd.concat([class_0,class_1])
```

```python
df2
```

```python
df_minaroty = df2[df2["target"]==1]
df_majority = df2[df2["target"]==0]
```

```python
df_majority_upsample = resample(df_majority,replace=True,n_samples=len(df_minaroty),random_state=42)
df_majority_upsample
```
