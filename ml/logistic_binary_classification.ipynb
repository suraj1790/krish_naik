### **Logistic regression in binary class classification problem**

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt


```

```python
from sklearn.datasets import make_classification
X,y = make_classification(n_classes=2,n_features=10,n_samples=1000,random_state=42)
```

```python
X
```

```python
X = pd.DataFrame(X)
```

```python

X.head()
```

```python
plt.plot(X)
```

```python
sns.boxplot(X)
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=42)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
```

```python
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_test)
```

```python
from sklearn.metrics import precision_score,accuracy_score,confusion_matrix,classification_report
```

```python
print("accuracy_score : ",accuracy_score(y_test,y_pred))
print("precision_score : ",precision_score(y_test,y_pred))
print("confusion_matrix : ",confusion_matrix(y_test,y_pred))
```

```python
print(classification_report(y_test,y_pred))
```

### Hyperparameter tuning

```python
penalti = ["l1","l2","elasticnet"]
c = [0.1,10,100,0.01,20,30,50]
solver = ['lbfgs', 'liblinear', 'saga']
```

```python
params = {"penalty":["l1","l2","elasticnet"],"C":[0.1,10.0,100.0,0.01,20.0,30.0,50.0],"solver":['lbfgs', 'liblinear', 'saga']}
model2 = LogisticRegression()
```

```python
from sklearn.model_selection import GridSearchCV,RandomizedSearchCV
grid = GridSearchCV(estimator=model2,param_grid=params)
```

```python
grid.fit(x_train,y_train)
```

```python
y_pred = grid.predict(x_test)
print(accuracy_score(y_test,y_pred))
```
